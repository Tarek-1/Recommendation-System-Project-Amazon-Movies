# Cleaning

```python
df_copy = df.copy()
df_copy.head()
```

```python
#print the shape of the data
df_copy.shape
```

```python
# replace spaces and symbols with underscores
df_copy.columns = (
    df_copy.columns
    .str.strip()                          # Trim leading/trailing whitespace
    .str.lower()                          # Convert to lowercase
    .str.replace(' ', '_')                # Replace spaces
    .str.replace('-', '_')                # Replace spaces
    .str.replace('/', '_')                # Replace slashes
    .str.replace(r'[\(\)\$]', '', regex=True)  # Remove parentheses and dollar signs
    .str.replace('__+', '_', regex=True)  # Collapse multiple underscores
    .str.strip('_')                       # Remove leading/trailing underscores
)
df_copy.columns
```

```python
# First, create a new column by combining 'country' and 'year'
df_copy['country_year_check'] = df_copy['country'].astype(str) + df_copy['year'].astype(str)

# Now compare it with the existing 'country_year' column
if 'country_year' in df_copy.columns:
    match = (df_copy['country_year_check'] == df_copy['country_year']).all()
    print("Are 'country_year' and combined 'country' + 'year' identical?:", match)

    # Optionally, show mismatches if they exist
    mismatches = df_copy[df_copy['country_year_check'] != df_copy['country_year']]
    print("Number of mismatched rows:", len(mismatches))
else:
    print("'country_year' column not found.")
```

this means that the column is not useful and we will drop it

```python
df_copy.drop(['country_year','country_year_check'], axis =1, inplace= True)
df_copy.shape
```

```python
# Convert gdp_for_year to float (remove commas and convert type)
df_copy['gdp_for_year'] = df_copy['gdp_for_year'].str.replace(',', '').astype(float)
```

```python
# Group and sum suicides
grouped = df_copy.groupby(['year', 'age'])['suicides_no'].sum().reset_index()

# Normalize per year (rate within each year)
grouped['norm_suicides'] = grouped.groupby('year')['suicides_no'].transform(lambda x: x / x.sum())

# Create pivot tables for raw and normalized
pivot_raw = grouped.pivot(index='age', columns='year', values='suicides_no')
pivot_norm = grouped.pivot(index='age', columns='year', values='norm_suicides')

# Set axis labels
pivot_raw.index.name = pivot_norm.index.name = "Age Group"
pivot_raw.columns.name = pivot_norm.columns.name = "Year"

# custom hover text for raw
hover_raw = [
    [f"Year: {year}<br>Age Group: {age}<br>Suicides: {int(val):,}" if pd.notna(val) else ""
     for year, val in zip(pivot_raw.columns, row)]
    for age, row in zip(pivot_raw.index, pivot_raw.values)
]

# Prepare custom hover text for normalized
hover_norm = [
    [f"Year: {year}<br>Age Group: {age}<br>Normalized Rate: {val:.2%}" if pd.notna(val) else ""
     for year, val in zip(pivot_norm.columns, row)]
    for age, row in zip(pivot_norm.index, pivot_norm.values)
]

fig = go.Figure()

# Raw heatmap trace
fig.add_trace(
    go.Heatmap(
        z=pivot_raw.values,
        x=pivot_raw.columns.astype(str),
        y=pivot_raw.index.astype(str),
        colorscale='Greys',
        colorbar=dict(title="Suicides", tickformat=","),
        text=hover_raw,
        hoverinfo="text",
        visible=True,
        name="Raw Count"
    )
)

# Normalized heatmap trace
fig.add_trace(
    go.Heatmap(
        z=pivot_norm.values,
        x=pivot_norm.columns.astype(str),
        y=pivot_norm.index.astype(str),
        colorscale='Greys',
        colorbar=dict(title="Normalized Rate"),
        text=hover_norm,
        hoverinfo="text",
        visible=False,
        name="Normalized Rate"
    )
)

# Dropdown to toggle between raw and normalized
fig.update_layout(
    title="Number of Suicides by Year and Age Group",
    xaxis=dict(title="Year", type="category", tickangle=45),
    yaxis=dict(title="Age Group", type="category"),
    font=dict(size=12),
    margin=dict(l=40, r=40, t=60, b=40),
    updatemenus=[
        dict(
            buttons=[
                dict(label="Raw Count",
                     method="update",
                     args=[{"visible": [True, False]},
                           {"title": "Number of Suicides by Year and Age Group",
                            "coloraxis": {"colorbar": dict(title="Suicides", tickformat=",")}}]),
                dict(label="Normalized Rate",
                     method="update",
                     args=[{"visible": [False, True]},
                           {"title": "Normalized Suicide Rate by Year and Age Group",
                            "coloraxis": {"colorbar": dict(title="Normalized Rate")}}]),
            ],
            direction="down",
            showactive=True,
            x=0.5,
            xanchor="center",
            y=1.15,
            yanchor="top"
        )
    ]
)

fig.show()
```

###  Heatmap: Number of Suicides by Year and Age Group (Raw Count)

This heatmap shows the **absolute number of suicides** per age group across different years.

####  Key Insights:
- **Darker shades** represent **higher numbers of suicides**.
- The **35–54 age group** consistently has the darkest regions, indicating the highest total suicides.
- After **2015**, there is a **sharp drop** in intensity — especially in **2016**, which appears extremely light across all age groups.
  - This is due to **missing or incomplete data** for those years, not an actual drop in suicides.

---

###  Why Use Normalized Rates?

The post-2015 records contain **fewer total entries**, especially in **2016**, which skews the raw counts downward.

####  Purpose of Normalization:
- **Normalization was applied per year**, not per population, to **correct for inconsistent data volume across years**.
- This ensures each year contributes equally to visual patterns — allowing fair comparisons between years despite incomplete records.

####  Example:
- Without normalization, **2016 looks artificially low** simply because fewer records were logged.
- With normalization, we adjust each year’s values to a **comparable scale**, making **patterns over time** easier to spot.
